```python
TSV=pd.read_table('gdc_sample_sheet.2022-09-20.Tum.tsv')
UID=pd.read_table('gdc_manifest_20220920_155722.txt')
PrimaryTumour=TSV[TSV['Sample Type']=='Primary Tumor']
PrimaryTumour=TSV[TSV['Sample Type']=='Primary Tumor']
MatchedCon=TSV[TSV['Sample Type']=='Solid Tissue Normal'].sort_values(by='Case ID')
MatchedCase=PrimaryTumour.loc[PrimaryTumour['Case ID'].isin(MatchedCon['Case ID'])].sort_values(by='Case ID')
```

```python
H_ID=MatchedCon['File Name']
T_ID=MatchedCase['File Name']
con=ALL[H_ID]
tumour=ALL[T_ID]
con.columns=MatchedCon['Case ID']
con.index=ALL['gene_id']
tumour.columns=MatchedCase['Case ID']
tumour.index=ALL['gene_id']
```

```python
ALL=pd.concat([con, tumour], axis=1)
data=ALL[(~ALL.astype('bool')).mean(axis=1)<0.80]
con=con.loc[data.index]
hcc=tumour.loc[data.index]
```

```python
ALL['TCGA-BC-A10Q']
```

```python
Con=np.log2(con+1) 
HCC=np.log2(hcc+1) 
meanTPM_con=Con.mean(axis=1)
meanTPM_hcc=HCC.mean(axis=1)
```

```python
x=HCC.values-Con.values
log2Fc_raw=pd.DataFrame(x)
log2Fc_raw.columns=Con.columns
Log2Fc=log2Fc_raw.mean(axis=1)
fc=2**Log2Fc
Fc=[]
for i in fc:
    if (i<1):
        Fc.append(-1/i)
    else:
        Fc.append(i)
Fc
```

```python
Log2Fc=pd.DataFrame(Log2Fc)
Log2Fc['Gene']=Con.index
```

```python
result=[]
for i in Con.index:
    x1 = Con.loc[i]
    x2 = HCC.loc[i]
    ttest_result=scipy.stats.ttest_rel(x1,x2)
    result.append(
        {
            'Gene': i,
            'T': ttest_result.statistic,
            'p value':  ttest_result.pvalue
        })


```

```python
Result=pd.DataFrame(result)

p_adjusted=multi.fdrcorrection(Result['p value'], alpha=0.05, method='indep', is_sorted=False)
Result['p adjusted']=p_adjusted[1]
Result['Fc']=Fc
Result['Average Normalized Values in Adjacent Non-Tumour']=np.array(meanTPM_con)
Result['Average Normalized Values in Tumour']=np.array(meanTPM_hcc)
Summary = Result[['Gene','Average Normalized Values in Tumour','Average Normalized Values in Adjacent Non-Tumour','Fc','p value','p adjusted']]
Summary
```
